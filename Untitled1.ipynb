# **Read Excel and Operations with Nan**

```python
# Load Excel file
df = pd.read_excel('basic.xlsx')
```

```python
# Display the DataFrame
print(df.head(6))
```

```python
print("Original DataFrame:")
print(df)
```

```python
# Specify the index of the row to be dropped
index_to_check = 3

# Check if the row at the specified index contains 'None' in both columns
# Check if the row at the specified index contains NaN in both columns
if pd.isna(df.loc[index_to_check, 'First Name']) and pd.isna(df.loc[index_to_check, 'Last Name']):
    df = df.drop(index=index_to_check)
```

# **Read Excel from columns in Middle**

```python
df2 = pd.read_excel('/content/Book1.xlsx', usecols='C:D', skiprows=1, nrows=13)
```

```python
df2
```

```python
# Specify the index of the row to be dropped
index_to_check = 6

# Check if the row at the specified index contains 'None' in both columns
# Check if the row at the specified index contains NaN in both columns
if pd.isna(df2.loc[index_to_check, 'First Name']) and pd.isna(df2.loc[index_to_check, 'Last Name']):
    df2 = df2.drop(index=index_to_check)
```

```python
df2
```

```python
# Iterate over each row in the DataFrame
for index, row in df2.iterrows():
  if row['Last Name'] in ['Gala', 'Morbia', 'Modi']:
    df2.loc[index, 'First Name'] = 'Rohit'  # Update 'First Name' for matching rows
  elif row['Last Name'] in ['Kale', 'Sawant', 'Sabnis']:
    df2.loc[index, 'First Name'] = 'Ankit'
  else:
    df2.loc[index, 'First Name'] = 'Mohit'
```
